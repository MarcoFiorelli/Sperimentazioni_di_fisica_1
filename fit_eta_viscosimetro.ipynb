# Eta (non newtoniano)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.750, 1.000, 1.588, 1.084, 2.381, 2.788, 3.175, 3.572])
y = np.array([7.75, 7.71, 7.37, 6.90, 6.34, 5.74, 4.73, 3.06])
s = np.array([0.13, 0.08,	0.06, 0.04, 0.03, 0.03, 0.03, 0.03])
```

```python
def retta(t, a, b):
  return a*t+b
```

```python
from types import FunctionType
# Definisci i parametri per la retta orizzontale
a = 5.705 # Modifica questo valore per spostare la retta verticalmente
b = 0 # Questo deve essere 0 per una retta orizzontale

# Genera i valori y per la retta orizzontale usando la funzione 'retta'
y_orizzontale = retta(x, b, a)

# Ridisegna il grafico con la nuova retta
plt.figure(figsize=(12, 7), dpi=200)

plt.xlabel('Raggi [mm]', labelpad=10, fontsize=15)
plt.ylabel('Viscosità [Pa s]', labelpad=10, fontsize=15)
plt.grid(which='major', linewidth=0.8)
plt.minorticks_on()

# Plot di tutti i punti sperimentali con il colore di default
plt.errorbar(x, y, yerr=s, fmt='o')

# Plot della nuova retta orizzontale
plt.plot(x, y_orizzontale, color='red', label=f'Retta Orizzontale (y={a:.3f})')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```

# Eta (newtoniano)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.750, 1.000, 1.588, 1.084, 2.381, 2.788, 3.175, 3.572])
y = np.array([1.18, 1.19, 1.19, 1.19, 1.17, 1.20, 1.20, 1.23])
s = np.array([0.02, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01])
```

```python
def retta(t, a, b):
  return a*t+b
```

```python
from types import FunctionType
# Definisci i parametri per la retta orizzontale
a = 1.1946667 # Modifica questo valore per spostare la retta verticalmente
b = 0 # Questo deve essere 0 per una retta orizzontale

# Genera i valori y per la retta orizzontale usando la funzione 'retta'
y_orizzontale = retta(x, b, a)

# Ridisegna il grafico con la nuova retta
plt.figure(figsize=(12, 7), dpi=200)

plt.xlabel('Raggi [mm]', labelpad=10, fontsize=15)
plt.ylabel('Viscosità [Pa s]', labelpad=10, fontsize=15)
plt.grid(which='major', linewidth=0.8)
plt.minorticks_on()

# Plot di tutti i punti sperimentali con il colore di default
plt.errorbar(x, y, yerr=s, fmt='o')

outlier = [4, 7] # Indici 0-based per il 5° e l'8° punto
plt.errorbar(x[outlier], y[outlier], yerr=s[outlier], fmt='o', color='orange', markersize=8)

# Plot della nuova retta orizzontale
plt.plot(x, y_orizzontale, color='red', label=f'Retta Orizzontale (y={a:.3f})')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```
