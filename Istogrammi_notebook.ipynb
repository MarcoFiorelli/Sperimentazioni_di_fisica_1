# Istogramma (F=500)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

```

```python
dati_al=np.array([43, 45, 45, 45,
45, 46, 45, 45, 46, 44, 44, 43,
43, 42, 43, 43, 44, 45, 45, 43])

bins_al=np.array([41.5, 42.5, 43.5, 44.5, 45.5, 46.5])

dati_ac=np.array([47, 45, 46, 45,
45, 44, 45, 43.5, 45, 45, 46, 47,
43, 47, 46, 46, 47, 46, 47, 47])

bins_ac=np.array([42.5, 43.5, 44.5, 45.5, 46.5, 47.5])
```

```python
media_al=np.mean(dati_al)
dev_std_al=np.std(dati_al, ddof=1)

media_ac=np.mean(dati_ac)
dev_std_ac=np.std(dati_ac, ddof=1)

print("media_al=", media_al)
print("dev_std_al=", dev_std_al)

print("media_ac", media_ac)
print("dev_std_ac=", dev_std_ac)
```

```python
plt.hist(dati_al, bins=bins_al, alpha=0.3,
         color="red")
plt.hist(dati_ac, bins=bins_ac, alpha=0.3,
         color="blue")

plt.axvline(media_al, color="red", linestyle="dashed",
            linewidth=1, label="media allungamento")
plt.axvline(media_ac, color="blue", linestyle="dashed",
            linewidth=1, label="media accorciamento")

plt.grid(which='major', linewidth=0.5)
plt.minorticks_on()

ax=plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))

plt.xlabel("variazione di lunghezza in modulo [$10^{-5}$ m]")
plt.ylabel("Frequenza assoluta")

plt.legend(
    loc="upper left",
    fontsize=8,
    frameon=True,
)

plt.show()
```
